```python
import matplotlib.pyplot as plt
import numpy as np

# 격자 생성
grid = np.zeros((4, 6))

# 다익스트라 알고리즘의 경로를 저장할 리스트
fwa_path = [(0, 0), (1, 0), (2, 0), (3, 0), (3, 1), (3, 2), (2, 2), (1, 2), (0, 2), (0, 3), (0, 4), (1, 4), (2, 4), (2, 5), (3, 5)]
```

```python
# 다익스트라 알고리즘 경로 그리기
plt.figure(figsize=(5, 5))
plt.title('Dijkstra Path')
plt.imshow(grid, cmap='binary')
plt.scatter(*zip(*fwa_path), color='blue')
plt.plot(*zip(*fwa_path), color='blue')
plt.xticks(range(4))
plt.yticks(range(6))
plt.grid(True)
plt.show()
```
